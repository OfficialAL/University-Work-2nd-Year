## **5. Impute the Building_Class_Category with a Lable Encoder (Hint check Section 4 Imputing Categorical Data in Lab example)**

```python
# Check the values of the Building_Class_Category
df['Building Class Category'].unique()
df['Building Class Category'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd

# Create a copy to preserve original data
df['Building Class Category_Original'] = df['Building Class Category'].copy()

# Initialize encoder
le = LabelEncoder()

# Handle missing values
mask = df['Building Class Category'].notna()
encoded_values = le.fit_transform(df.loc[mask, 'Building Class Category'])
df.loc[mask, 'Building Class Category_Encoded'] = pd.Series(encoded_values, index=df.index[mask])


df[['Building Class Category', 'Building Class Category_Encoded']].head(10)
```
